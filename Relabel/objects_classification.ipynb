```python
def find_added_stillin_removed(current,last):
    added = []
    stillin = []
    removed = []
    for i in current:
        if i in last:
            stillin.append(i)
        else:
            added.append(i)
    for i in last:
        if i not in current:
            removed.append(i)
    return added,stillin,removed

current = [1,2,3,4,5]
last = [0,3,4]
find_added_stillin_removed(current,last)
```

```python
last_ids = []
rst = {}

for _ in range(len(info)):
    frame = info[info.FrameIdx == _]
    
    if _ == 0:
        for obj in frame.iterrows():
            objIdx,objId,className = obj[1].ObjIdx,obj[1].Obj_Id,obj[1].Class_Name
            last_ids.append(objId)
            rst[objId] = [{objId:[className]}]
        print(rst,'\n')
        continue
        
    current_info = {}    
    for obj in frame.iterrows():
        objIdx,objId,className = obj[1].ObjIdx,obj[1].Obj_Id,obj[1].Class_Name
        current_info[objId] = className
    current_ids = current_info.keys()
    #print('last\n',last_ids)
    #print('current\n',current_ids)
    added,stillin = find_added_stillin(current_ids,last_ids)
    #print('added,stillin\n',added,stillin,'\n')
    last_ids = current_ids  
    for i in stillin:
        if current_info[i] not in rst[i][-1][i]:
            rst[i][-1][i].append(current_info[i])
            #print('updated',i,current_info[i])
        #print(rst)
    for i in added:
        #print(i,current_info[i])
        if i not in rst:
            rst[i] = [{i:[current_info[i]]}]
        else:
            rst[i].append({i:[current_info[i]]})
    #print(rst,'\n==========================')
print(rst)
```

```python
huanchong = {}
last_ids = []
rst = []

for _ in range(len(info)):
    frame = info[info.FrameIdx == _]
    
    current_info = {}    
    for obj in frame.iterrows():
        objIdx,objId,className = obj[1].ObjIdx,obj[1].Obj_Id,obj[1].Class_Name
        current_info[objId] = className
    current_ids = current_info.keys()
    #print('last\n',last_ids)
    #print('current\n',current_ids)
    added,stillin,removed = find_added_stillin_removed(current_ids,last_ids)
    #print('added',added)
    #print('stillin',stillin)
    #print('removed',removed,'\n')
    last_ids = current_ids 
    for i in added:
        huanchong[i] = [current_info[i]]
    for i in stillin:
        if current_info[i] not in huanchong[i]:
            huanchong[i].append(current_info[i])
            #print('updated',i,current_info[i])
    for i in removed:
        tmp = {i:huanchong[i]}
        del huanchong[i]
        rst.append(tmp)
        #print('move to rst',tmp)
    #print('\nhuanchong',huanchong)
    #print('rst',rst,'\n====================================================')
print(rst)
```
